```python
import pandas as pd
import numpy as np
from fbprophet import Prophet

df = pd.read_csv('../assets/test.csv')
df['y'] = np.log(df['y'])
df['cap'] = 10.893572
df.head()

m = Prophet(growth='logistic')
m.fit(df)

future = m.make_future_dataframe(periods=100, freq='D')
future['cap'] = 10.893572
fcst = m.predict(future)
m.plot(fcst);
```
